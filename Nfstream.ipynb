```python
def ip_to_int(ip):
    try:
        ip_obj = ipaddress.ip_address(ip)
        if isinstance(ip_obj, ipaddress.IPv4Address):
            ip_int = int.from_bytes(socket.inet_aton(ip), 'big')
            return np.uint32(ip_int)
        elif isinstance(ip_obj, ipaddress.IPv6Address):
            return int(ip_obj)
    except ValueError as e:
        print(f"Invalid IP address {ip}: {e}")
        return np.nan

# Assuming df_Final is your DataFrame
df_Final["src_ip"] = df_Final['src_ip'].apply(ip_to_int)
df_Final["dst_ip"] = df_Final['dst_ip'].apply(ip_to_int)
```

```python
def mac_to_int(mac_str):
    # Remove colons and convert to integer
    try:
        return int(mac_str.replace(':', ''), 16)
    except ValueError:
        # Handle the case where mac_str is not a valid MAC address
        return None
# Assuming df is your DataFrame and 'src_mac' is the column with MAC addresses
df_Final['src_mac_normalized'] = df_Final['src_mac'].apply(mac_to_int)
# Check the result
# print(df_Final.head())
df_Final['src_mac_normalized']
```

### Split columns to x and y (features,target)

```python
X = df_Final[["src_ip",	"src_mac","src_oui","src_port","dst_ip","dst_mac","dst_oui","dst_port","protocol","ip_version","vlan_id","tunnel_id","bidirectional_first_seen_ms","bidirectional_last_seen_ms","bidirectional_duration_ms","bidirectional_packets","bidirectional_bytes","src2dst_first_seen_ms","src2dst_last_seen_ms","src2dst_duration_ms","src2dst_packets","src2dst_bytes","dst2src_first_seen_ms","dst2src_last_seen_ms",	"dst2src_duration_ms","dst2src_packets","dst2src_bytes"]]
# X = df_Final[["src_ip","dst_ip",""]]
# X_new = dff[["src_ip","dst_ip", "src_port","bidirectional_packets", "bidirectional_bytes"]
y = df_Final[['application_name', 'application_category_name']]
# y = df["application_category_name"].apply(lambda x: 1 if 'SocialNetwork' in x else 0) 
```
